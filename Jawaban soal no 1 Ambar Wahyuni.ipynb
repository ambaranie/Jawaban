```python
# Nomor 2 
# jumlah karyawan, jumlah kantor & lokasi negaranya, jumlah barang yang dijual, 
# total stok barang & jumlah vendor yang menjadi partner Anda.

emp = pd.read_sql("select count(distinct employeeNumber) as Employees from employees e", mydb)
off = pd.read_sql("select count(distinct officeCode) as Office, count(distinct country) as Country from offices o", mydb)
prod = pd.read_sql("select count(distinct productCode) as Product, sum(quantityInStock) as StockProduct, count(distinct productVendor) as Vendors from products p", mydb)

df = pd.DataFrame(emp['Employees'])
df['Office']=off['Office']
df['Country']=off['Country']
df['Product']=prod['Product']
df['StockProduct']=prod['StockProduct']
df['Vendors'] = prod['Vendors']
df
```

```python
# Nomor 3
# Rumuskan single query yang dapat menampilkan harga produk terendah & tertinggi untuk masing-masing kategori.
kategori = pd.read_sql("""select productLine,  min(buyPrice) as minPrice, max(buyPrice) as maxPrice 
from products group by productLine""", mydb)
kategori
```

```python
#Nomor 4
# Rumuskan single query yang dapat menampilkan daftar 10 customer paling royal (paling banyak mendatangkan uang bagi kita), 
# yang total nominaltransaksinya paling tinggi. 
# Data yang ditampilkan adalah nama customer, kota& negara asal, beserta total uang yang dihabiskan di produk kita.

transaksi = pd.read_sql("""select customerName, city, country, sum(amount) as total 
from customers s join payments p on p.customerNumber = s.customerNumber 
group by customerName  
order by total desc 
limit 10""", mydb)
transaksi
```
